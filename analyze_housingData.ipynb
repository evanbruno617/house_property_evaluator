```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os
from sklearn.metrics import r2_score
import numpy as np
```

```python
# read in median prices
housing = pd.read_csv('MedianPricesExisting.csv')
housing.head()
```

```python
# read in median time

time = pd.read_csv('MedianTime.csv')
time.head()
```

```python
# get list of counties only

counties = housing.columns.tolist()[1:-4]
```

```python
# remove county that does not match
counties.remove('S.F. Bay Area')
```

```python
# get ratio of median prices to time on market

final_df = pd.DataFrame()

counties_df = {}

for county in counties:

    transfer_county = housing[[f'{county}', 'Year']]
    
    transfer_time = time[[f'{county}', 'Year']]
    
    data_df = pd.merge(transfer_county, transfer_time, on='Year')
    
    data_df['Ratio'] = data_df[f'{county}_x'] / data_df[f'{county}_y']
    
    
    counties_df[county] = data_df['Ratio'].tolist()
    
    
    
```

```python
# make into dataframe

ratio_df = pd.DataFrame(counties_df)
```

```python
ratio_df['Year'] = housing['Year'].tolist()
```

```python
#export as csv
ratio_df.to_csv('Ratio_Change.csv')
```
